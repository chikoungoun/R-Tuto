```r
library(ggplot2)
library(palmerpenguins)
```

```r
data(penguins)
```

```r
head(penguins)
```

**color :** helps setting colors to differentiate the dots

```r
ggplot(data = penguins) + geom_point(mapping = aes(x=flipper_length_mm, y=body_mass_g,color = species))
```

**shape :** assign different shapes to differentiate the species

```r
ggplot(data = penguins) + geom_point(mapping = aes(x=flipper_length_mm, y=body_mass_g,shape = species))
```

**size :** assign different sizes to differentiate the species

```r
ggplot(data = penguins) + geom_point(mapping = aes(x=flipper_length_mm, y=body_mass_g,size=species))
```

**combine the aesthetics**

```r
ggplot(data = penguins) + geom_point(mapping = aes(x=flipper_length_mm, y=body_mass_g,shape = species,color=species,size=species))
```

**alpha :** assign different alpha (opacity) to differentiate the species

```r
ggplot(data = penguins) + geom_point(mapping = aes(x=flipper_length_mm, y=body_mass_g,alpha = species))
```

* Change the color of all the data points

```r
ggplot(data = penguins) + geom_point(mapping = aes(x=flipper_length_mm, y=body_mass_g),color="purple")
```
