* Caption

```r
ggplot(data=penguins)+
geom_point(mapping = aes(x=flipper_length_mm,y=body_mass_g,color=species))+
labs(title = "Palmer Penguins: Body Mass vs. Flipper Length",subtitle = "Sample of 3 penguins Species",caption = "Data collected by Dr. Kristen Gorman")
```

* Add annotation in the plot

```r
ggplot(data=penguins)+
geom_point(mapping = aes(x=flipper_length_mm,y=body_mass_g,color=species))+
labs(title = "Palmer Penguins: Body Mass vs. Flipper Length",subtitle = "Sample of 3 penguins Species",caption = "Data collected by Dr. Kristen Gorman")+
annotate("text",x=220,y=3500,label=" The Gentoos are the largest")
```

* Change the color, angle and size of the annotation

```r
ggplot(data=penguins)+
geom_point(mapping = aes(x=flipper_length_mm,y=body_mass_g,color=species))+
labs(title = "Palmer Penguins: Body Mass vs. Flipper Length",subtitle = "Sample of 3 penguins Species",caption = "Data collected by Dr. Kristen Gorman")+
annotate("text",x=220,y=3500,label=" The Gentoos are the largest",color="purple",fontface="bold",angle=25,size=4.5)
```

* Annotations with variables to make the code less dense

```r
p <- ggplot(data=penguins)+
geom_point(mapping = aes(x=flipper_length_mm,y=body_mass_g,color=species))+
labs(title = "Palmer Penguins: Body Mass vs. Flipper Length",subtitle = "Sample of 3 penguins Species",caption = "Data collected by Dr. Kristen Gorman")
```

```r
p+annotate("text",x=220,y=3500,label=" The Gentoos are the largest",color="purple",fontface="bold",angle=25,size=4.5)
```
